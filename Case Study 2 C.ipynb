#### Key issues
#### Build a system where when a user enters Reference ID it is encryptedso that hackers cannot view the mapping of Reference ID and fingerprint

#### 1.Read the input from the command line –Reference ID
#### 2.Check for validity –it should be 12 digits and allows fornumber and alphabet
#### 3.Encrypt the Reference ID and print it for reference

```python
from cryptography.fernet import Fernet

rid = input("Enter Refernece ID :")

if (len(rid) == 12 and rid.isalnum()) == False:
    print("it should be 12 digits and allows only number and alphabet")
else:
    key = Fernet.generate_key()
    cipher_suite = Fernet(key)
    encoded_text = cipher_suite.encrypt(rid.encode())
    decoded_text = cipher_suite.decrypt(encoded_text)
    print(decoded_text.decode())
```

#### 1.Allow some special characters in ReferenceID
#### 2.Give the option for decryption to the user 

```python
from cryptography.fernet import Fernet
import re

rid = input("Enter Refernece ID :")

if not (len(rid) == 12):
    print("it should be 12 digits")
else:
    if not (re.search("[#$]",rid) and (re.search("[0-9]",rid) or re.search("[a-z]",rid))):
        print("allows only number and alphabet")
    else:
        key = Fernet.generate_key()
        cipher_suite = Fernet(key)   
        encoded_text = cipher_suite.encrypt(rid.encode())
        choice = input("Key Encrypted want to decrypt ? (y/n): ")
        if choice.upper() == 'Y':
            decoded_text = cipher_suite.decrypt(encoded_text)
            print(decoded_text.decode())
```
