### Write a program that will find factors of the given number and find whether the factor is even or odd.
#### Hint: Use Loop with if-else statements

```python
num = 24
for n in range(1,num+1):
    if(num%n == 0):
        if(n%2==0):
            print(f"{n} is even")
        else:
            print(f"{n} is odd")
```

### Write a program, which will find all the numbers between 1000 and 3000 (both included) such that each digit of a number is an even number. The numbers obtained should be printed in a comma-separatedsequence on a single line.
#### Hint: In the case of input data being supplied to the question, it should be assumed to be a console input.Divide each digit by2 and verify whether is it even or not

```python
def check_even(i):
    for c in str(i):
        if(int(c)%2!=0):
            return False
    return True

for i in range(1000,3001):
    if check_even(i) == True:
        print(i,end=',')
```

### Write a program that accepts a sentence and calculatesthe number of letters and digits.
#### Suppose the entered string is: Python0325
#### Then the output will be:
#### LETTERS: 6
#### DIGITS:4
##### Hint: Use built-in functions of string

```python
alpha,num=0,0
str1 = "Python0325"
for s in str1:
    if str(s).isalpha():
        alpha += 1
    if str(s).isdigit():
        num += 1
print("LETTERS:",alpha)
print("DIGITS:",num)
```

### Design a code thatwill find whether the given number is a Palindrome number or not.
#### Hint: Use built-in functions of string.

```python
num = "11211"
if num[::-1] == num:
    print("Pallindrome")
else:
    print("Not a Pallindrome")
```
